```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rc("figure", figsize=(8, 8))
```

```python
df = pd.read_csv (r'/home/jason/storage-diss/gridwatch.csv', index_col=1, parse_dates=True ) #year of 2021
del df['id']
#df.plot()
df = df.sum(axis = 1)
resample3 = df.resample('3H')
resample12 = df.resample('12H')
r = resample12.mean()
HourlyData = resample3.mean()
df['2021-1-01':'2021-1-31'].plot(label='5 minute interval')
HourlyData['2021-1-01':'2021-1-31'].plot(label='3 hour interval')
r['2021-1-01':'2021-1-31'].plot(label='12 hour interval')
plt.legend(loc="upper left")
plt.ylabel('Power Output (MW)')
plt.xlabel('Time Stamp')
```

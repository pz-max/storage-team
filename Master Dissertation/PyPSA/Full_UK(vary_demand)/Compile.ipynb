```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#constant demand

X = ['5 minute', '30 minute','1 hour','3 hour','6 hour','12 hour']
Battery = [85.44779555456911,90.2522077064652,95.68197485703769,114.29398934371655,132.73091955708193,139.6982804903761]
H2 = [32.034484277353904,36.29395733422566,39.450362038418874,36.27298506926988,29.558,29.558]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Battery, 0.4, label = 'Battery')
plt.bar(X_axis + 0.2, H2, 0.4, label = 'H2')
  
plt.xticks(X_axis, X)
plt.xlabel("Time series intervals")
plt.ylabel("Objective Cost ($ billion)")
plt.title("Comparison of objective cost optimisations from different time series intervals \n for supply and demand data for the month of January 2021")
plt.legend()
plt.show()
```
